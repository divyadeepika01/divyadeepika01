```python
# line plot (we can find the relationship between categorical values also here in seaborn)
# dense curve tells us how much values are there in that area
sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species")
plt.show()
```

```python
# observations:
# 1. sepal_length of setosa ranges between 4.2 to 5.7 
# 2. sepal length of versicolor is ranges between 4.9 to 7.0 
# 3. virginica sepal length ranges between 4.8 to 7.9
# 4. after 7.0 the sepal length of versicolor and virginica not varying simultaniously.
# 5. before 4.3 there is no iris flower.
```

```python
sns.lineplot(x="sepal_length",y="species",data=df)
plt.show()
```

```python
sns.lineplot(x="petal_length",y="species",data=df)
plt.show()
```

```python
sns.lineplot(x="species",y="petal_width",data=df)
plt.show()
```

```python
# observations :
# 1. setosa's petal width increases from 0.25 to 1.26    (min. width is of setosa iris flower)
# 2. 
```

# Scatter plot 
* between two continuous variables.

```python
sns.scatterplot(x="petal_length",y="petal_width",data=df,hue="species")
plt.show()
```

## Bar Plot
* by default giving the bins acc. to mean ,can change by "estimator=median,count" 
* x-axis --> categorical, y-axis--> numerical column 

```python
sns.barplot(x="species",y="petal_length",data=df)
plt.show()
```

# Count Plot
*  between categorical and numerical columns
* count the cetegories

```python
sns.countplot(x="species",data=df)
plt.show()
```

# Cat Plot
* almost like bar plot only

```python
sns.catplot(x="species",y="petal_length",kind="bar",data=df)
plt.show()
```

# Box Plot

```python
sns.boxplot(x="sepal_length",data=df)
plt.show()
```

```python
# Checking outliers of each category with respect to sepal width.
sns.boxplot(x="species",y="sepal_width",data=df)
plt.show()
```

# Violin Plot
* it is similar to the boxplot except that it provides a higher,
* more advanced visualisation and uses the kernel
* density estimation to give a better description about the data distribution.

```python
sns.violinplot(x="species",y="sepal_width",data=df)
plt.show()
```

# Strip Plot
* It basically creates a scatter plot based on the category.
* It is created using the stripplot()

```python
sns.stripplot(x="species",y="petal_width",data=df)
plt.show()
```

# Histogram
* A histogram is basically used to represent data provided in a form of some groups
* It is accurate method for the graphical representation of numerical distribution.

```python
sns.histplot(x="petal_width",data=df)
plt.show()
```

```python
sns.histplot(x="petal_width",data=df,hue="species")
plt.show()
```

```python
# kde (kernel density estimation) is approximation of histogram to density curve.
sns.histplot(x="sepal_length",data=df,hue="species",kde=True)
plt.show()
```

# Distplot
* is used basically for univariant set of observations
* and visualizes it trough a histogram
* i.e. only one observation and hence we choose one particular
* column of the dataset
** density means the probability of given plot to lie on the curve.

```python
sns.distplot(df.petal_width,color="yellowgreen")
plt.show()
```

```python
sns.pairplot(data=df,hue="species")
plt.plot()
```

3. petal_length and petal_width of setosa is min.



# Heap Map   : shows the highly correlated feature

* highly correlated if it cross the threshhold value(-0.9 or +0.9)
* on checking the pearsons coefficient go with the magnitude only not with the direction(+ or -)
* correlation scaling(-1 to +1) version of covariance(-infinity to +infinity). 

```python
# for ML model there should be no correlated feature with the variables because it increases the training time of model
# however it is ok for having highly correlated feature 
#  with the target (output) variable.
```
