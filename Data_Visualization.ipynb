# Line Plot

Use when we have continuous data
* to check any trend in the data we can use lineplot

```python
x=[1,8,9,5,3,4,7,5,6,8,12]
y=[1,4,3,6,7,3,6,7,8,9,13]
```

```python
plt.plot(x)
plt.xlabel("Index Location")
plt.ylabel("Data")
plt.title("Line plot of a single list")
plt.show()
```

```python
#  to hide the memory location of the plot we use plt.show() ---->
plt.plot(x,linestyle='--',marker='*',color="r")           # marker is used to highlight the datapoints 
plt.show()
```

```python
# Plotting two continuous variables using line plot:-

plt.plot(x,y)    # relationship
plt.show()
```

```python
#  2 variables on same plot to compare different variables:-
plt.plot(x,marker="o")
plt.plot(y,marker="*")
plt.show()
```

# Bar plot

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
 
 * x-axis= data
 * y-axis=count

```python
#  creating dataset :-
data = {"C":15,"C++":35,"Java":20,"Python":30,"Data Science":40}
courses=list(data.keys())
students_counts=list(data.values())
  
fig = plt.figure(figsize = (8, 5))
 
# creating the bar plot
plt.bar(courses,students_counts,color="green",width=0.5)
 
plt.xlabel("Courses")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

# Pie Plot
use it for categorical data and discrete

```python
pielist=[1,2,3,4,8] # weightage of observations
plt.pie(pielist,labels=["Mango","Banana","Apple","Grapes","Orange"],autopct="%.2f")
plt.show()
```

```python
import math
x=np.arange(0,math.pi*2,0.05)
x
```

```python
y=np.sin(x)
y
```

```python
# Creating blank or empty canvas :-
fig=plt.figure()    # to create empty canvas
# ADDING AXES-->
ax=fig.add_axes([0.5,0.5,0.5,0.5])       #50% of the canvas
# LABELING THE AXES -->
ax.set_title("Sine wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
ax.plot(x,y)
plt.show()
```

```python
# Creating blank or empty canvas :-
fig=plt.figure()    # to create empty canvas
# ADDING AXES-->
ax=fig.add_axes([0.8,0.8,0.5,0.5])       #50% of the canvas
# LABELING THE AXES -->
ax.set_title("Sine wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Sine")
ax.plot(x,y)
plt.show()
```

```python
z=np.cos(x)
z
```

```python
fig=plt.figure()    

ax=fig.add_axes([0.5,0.5,0.8,0.8])       

ax.set_title("Cos wave")
ax.set_xlabel("Angle")
ax.set_ylabel("Cosine")
ax.plot(x,z)
plt.show()
```

# Sub - Plots -->

```python
# Commom layouts plot

fig,a=plt.subplots(2,2)
x=np.arange(1,6)

a[0][0].plot(x,x*x)
a[0][0].set_title("line plot")

a[0][1].hist(x)
a[0][1].set_title("Histogram")

a[1][0].plot(x,np.sqrt(x))
a[1][0].set_title("Square root")

a[1][1].plot(x,np.log(x))
a[1][1].set_title("Log")

plt.show()
```
